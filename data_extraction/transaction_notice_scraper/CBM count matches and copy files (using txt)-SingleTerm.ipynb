```python
# Step 2: (can be skipped if text files exists) Extract text from PDFs and save it to text
extracted_pdfs = []
                  
for file_name in file_names: 
    text = str(textract.process(path+file_name))
    # each extracted text is saved in .txt with same file name
    with open(path+file_name+'.txt','w+') as f:
        f.write(text)
```

```python
# Step 4: Copy files with more than 0 keywords to a new folder
from shutil import copyfile

# read_path = '/Users/Marcus/Documents/GitHub/LUX/10/'

read_path = '/Users/Marcus/Google Drive/Country files - transaction raw documents/Luxembourg/CBM merger files/2019/11-2019/pdf/'
# change this path accordingly to read the pdf's from their location

# write_path = '/Users/Marcus/Documents/GitHub/LUX/10/results/'

write_path = '/Users/Marcus/Google Drive/Country files - transaction raw documents/Luxembourg/CBM merger files/2019/11-2019/results/'

# change path accordingly for writing and storing the sorted pdf's. 

# The subfolder 'pdf's' needs to be created manually beforehand!

all_results = pickle.load(open(path+'all_search_results.pickle','rb'))
for file in all_results:
    search_terms = list(all_results[file].keys())
    break
    
print('Created folders for the following search terms: ', search_terms)

for term in search_terms:
    new_folder_name = term
    if os.path.exists(write_path+new_folder_name)==False:
        os.makedirs(write_path+new_folder_name)
    
    for file_name in all_results:
        
        match_quantity = all_results[file_name][term]
        file_name=file_name[:-4]
        if match_quantity>0:
            copyfile(read_path+file_name,write_path+new_folder_name+'/'+file_name)
            print('File ', file_name, ' moved to folder ', term)
```
