```python
import csv
import os  

opath = os.getcwd()  
print ("The current working directory is %s" % opath)
print()

def createDirectories(companyids, ctype):
    for companyid in companyids:
        path = opath+"\\"+ctype+"\\"+str(companyid)

        try:  
            os.mkdir(path)
        except OSError:  
            print ("Creation of the directory %s failed" % path)
        else:  
            print ("Successfully created the directory %s " % path)
            
companyIDs = []
 
with open('MCs/companies_annual_accounts_mcs.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        companyIDs.append(row[0])

print("List of Luxembourg company IDs from the registry:")
print("-------------------------------------------------")
print()
uniqueIDs = list(set(companyIDs))
print(uniqueIDs)
print()
print("Number of companies:", len(uniqueIDs))
print()
print("Creating directories for each company - these will be locations where we will store the annual account documents that we download from the registry...")
print()

createDirectories(uniqueIDs, "MCs")
```

```python
accompanyIDs = []
 
with open('ACs/companies_annual_accounts_acs.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        accompanyIDs.append(row[0])
        
print(accompanyIDs)
print(len(accompanyIDs))
acuniqueIDs = list(set(accompanyIDs))
print(acuniqueIDs)
print(len(acuniqueIDs))

finalacuniqueIDs = []

for acid in acuniqueIDs:
    if acid not in uniqueIDs:
        finalacuniqueIDs.append(acid)

print(finalacuniqueIDs)
print(len(finalacuniqueIDs))

        
createDirectories(finalacuniqueIDs, "ACs")
```
